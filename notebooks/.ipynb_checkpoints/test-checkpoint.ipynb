```python
import re
import argparse
import json
import pickle

from tqdm import tqdm

import numpy as np

import torch
import torch.nn.functional as F

from collections import Counter

from transformers import AutoModelForCausalLM, AutoTokenizer, set_seed

set_seed(42)

dataset_paths = {
    'aime': '/scratch/saydalie/llm_reasoning/data/aime_1983_2023/train.jsonl',
    'gsm8k': '/scratch/saydalie/llm_reasoning/data/gsm8k/train.jsonl',
    'math': '/scratch/saydalie/llm_reasoning/data/math/train.jsonl'
}

def load_model(model_name: str):
    # deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B: {"<think>": 151648, "</think>": 151649}
    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        torch_dtype="auto",
        device_map="auto"
    )
    tokenizer = AutoTokenizer.from_pretrained(model_name)

    return model, tokenizer
```

## Generate CoT

```python
def extract_solution_gsm8k(text: str) -> str:
    """
    Extracts the final solution value from a string where the solution
    is marked after '####'.
    """
    marker = "\n####"
    return text.split(marker, 1)[1].strip()

def generate_cot(prompt, tokenizer, model, max_new_tokens=8192):
    """
    Returns:
        - Prompt+CoT tokens
        - Prompt+CoT token-ids
        - CoT token-level entropies
        - CoT token-level </think> probabilities
        - CoT token-level </think> log-probabilities
        - prompt length
    """
    messages = [
        {"role": "user", "content": prompt}
    ]
    inputs = tokenizer.apply_chat_template(
        messages,
        add_generation_prompt=True,
        tokenize=True,
        return_dict=True,
        return_tensors="pt",
    ).to(model.device)
    
    # Generate output with logits
    output = model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        return_dict_in_generate=True,
        output_logits=True
    )
    
    # Get the prompt length
    input_and_output_ids = output.sequences
    input_len = inputs['input_ids'].shape[1]

    # Compute the entropy and </think> probability at for each token
    output_entropies = []
    output_eot_probs = []
    for logits in output.logits:
        probs = F.softmax(logits, dim=-1)           # [batch_size, vocab_size]
        log_probs = F.log_softmax(logits, dim=-1)   # [batch_size, vocab_size]
        entropy = -(probs * log_probs).sum(dim=-1)  # [batch_size]
        output_entropies.append(entropy)

        # </think> token's probability
        output_eot_prob = probs[:, 151649].item()
        output_eot_probs.append(output_eot_prob)
    
    # Stack to get shape [batch_size, output_len]
    output_entropies = torch.stack(output_entropies, dim=1)
    
    # Flatten tensors (assumes batch size = 1)
    input_and_output_token_ids = input_and_output_ids[0].tolist()
    output_entropies = output_entropies[0].tolist()

    # parsing thinking content
    try:
        # rindex finding 151649 (</think>)
        end_of_think_id = len(input_and_output_token_ids) - input_and_output_token_ids[::-1].index(151649) # corresponds to position id of </think> + 1
        input_and_cot_ids = input_and_output_token_ids[:end_of_think_id]
        cot_entropies = output_entropies[:end_of_think_id-input_len]
        cot_eot_probs = output_eot_probs[:end_of_think_id-input_len]
    except ValueError:
        input_and_cot_ids = input_and_output_token_ids
        cot_entropies = output_entropies
        cot_eot_probs = output_eot_probs
    
    # Detokenize each token individually
    input_and_cot_tokens = [tokenizer.decode([tid], skip_special_tokens=False) for tid in input_and_cot_ids]

    return input_and_cot_tokens, input_and_cot_ids, cot_entropies, cot_eot_probs, input_len

# ['aime', 'gsm8k', 'math']
dataset = 'aime'
# load the dataset
with open(dataset_paths[dataset], 'r') as f:
    data = [json.loads(line) for line in f]

prompts = []
for item in data:
    prompts.append(
        {
            'question': item['question'],
            'answer': item['answer']
        }
    )

model, tokenizer = load_model(model_name='deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B')

# TODO: Implement COT generation for the dataset
for item in prompts:
    prompt = item['question']
    # adhere to the suggested use
    prompt += ' Please reason step by step, and put your final answer within \\boxed{}'
    
    input_and_cot_tokens, input_and_cot_ids, cot_entropies, cot_eot_probs, input_len = generate_cot(
        prompt=prompt,
        tokenizer=tokenizer,
        model=model
    )
    # Save or process the generated COT as needed

    break
```

## Split into Chunks

```python
def chunk_cot(input_and_cot_tokens, input_and_cot_ids, cot_eot_probs, input_len, method):
    """
    Splits a CoT into chunks with 4 different methods:
        - "keywords": split at specific keywords ("wait", "alternatively").
        - "newlines": split at double-newline ("\n\n").
        - "uniform": split into (64) ~equal-sized chunks.
        - "eot_threshold": split at punctuation marks with </think> probability above a certain threshold.
    """

    assert method in ['keywords', 'eot_threshold', 'uniform', 'newlines']
    
    if method == 'keywords':
        keywords = ['wait', 'alternatively']
        indices = [i for i, token in enumerate(input_and_cot_tokens) if token.lower().strip() in keywords and i > input_len]
    
    elif method == 'newlines':
        delimiter = '\n\n'
        indices = [i for i, token in enumerate(input_and_cot_tokens) if delimiter in token and i > input_len]
    
    elif method == 'uniform':
        cot_len = len(input_and_cot_tokens) - input_len
        if cot_len < 128:
            num_chunks=4
        elif cot_len < 512:
            num_chunks=8
        elif cot_len < 1024:
            num_chunks=16
        elif cot_len < 2048:
            num_chunks=32
        else:
            num_chunks=64
        chunk_size = max(1, cot_len // num_chunks)
        indices = [input_len + (i + 1) * chunk_size for i in range(num_chunks - 1)]
    
    elif method == 'eot_threshold':
        punctuations=['.', '?', '!']
        punct_indices = []
        i = input_len
        while i < len(input_and_cot_tokens):
            if input_and_cot_tokens[i] in punctuations:
                j = i
                while j + 1 < len(input_and_cot_tokens) and input_and_cot_tokens[j+1] in punctuations:
                    j += 1
                if j < len(input_and_cot_tokens) - 1:
                    # take last punctuation in the run
                    # that is not the last token
                    punct_indices.append(j)
                i = j + 1
            else:
                i += 1
        
        if punct_indices:
            punct_indices = np.array(punct_indices)
            punct_eot_log_probs = np.log(np.array(cot_eot_probs)[punct_indices - input_len + 1])

            # threshold = mean+std of eot_log_prob of punctuations
            log_threshold = punct_eot_log_probs.mean() + punct_eot_log_probs.std()

            indices = punct_indices[np.where(punct_eot_log_probs > log_threshold)[0]].tolist()

        else:
            # when no punctuation found
            indices = []
        
    else:
        raise ValueError(f"Unknown method: {method}")

    # Add sentinel end indices
    if input_and_cot_tokens[-1] == '</think>':
        ends = indices + [len(input_and_cot_ids)-1]
    else:
        ends = indices + [len(input_and_cot_ids)]

    # Slice and filter out empty lists
    chunk_end_ids = []
    chunks_cumulative = []
    for end in ends:
        if end > 0 and input_and_cot_ids[:end]:
            chunk_end_ids.append(end)
            chunks_cumulative.append(input_and_cot_ids[:end])

    return chunk_end_ids, chunks_cumulative

def extract_boxed_solution(text: str):
    """
    Extract the answer enclosed in the last LaTeX-style \boxed{} from the given text.
    """
    matches = re.findall(r"\\boxed\{([^}]*)\}", text)
    return matches[-1].strip() if matches else None
```

```python
import re
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

from sympy import Rational, simplify
from sympy.core.sympify import SympifyError
from sympy.parsing.sympy_parser import (
    parse_expr,
    standard_transformations,
    implicit_multiplication_application,
)

def normalize_arrays(arrays):
    # Decide target length
    if not arrays:
        return [], []

    m = max(len(seq) for seq in arrays)

    # Common grid in [0, 1]
    common_x = np.linspace(0, 1, m)
    resampled = []

    for seq in arrays:
        n = len(seq)
        if n == 0:
            raise ValueError("Input arrays must not be empty.")
        # Single value → flat line
        if n == 1:
            resampled.append(np.full(m, seq[0], dtype=float))
            continue

        x = np.linspace(0, 1, n)
        y = np.asarray(seq, dtype=float)
        resampled.append(np.interp(common_x, x, y))

    return common_x, np.vstack(resampled).mean(axis=0)

def clean_math_answer(ans: str) -> str:
    """
    Clean and normalize a math answer string, similar to LightEval's approach,
    with support for implicit multiplication and symbolic simplification.
    
    Steps:
    - Strip whitespace and commas in numbers.
    - Remove stray $ or punctuation.
    - Parse with sympy (with implicit multiplication).
    - If numeric, reduce to rational form.
    - If symbolic, simplify and return canonical string.
    - Fallback: return normalized lowercase string.
    """
    if ans is None:
        return ""

    # Step 1: strip whitespace
    cleaned = ans.strip()
    cleaned = cleaned.replace(" ", "")

    # Step 2: remove commas in numbers
    cleaned = cleaned.replace(",", "")

    # Step 3: remove enclosing $ signs, units, or stray punctuation
    cleaned = cleaned.strip(" $")
    cleaned = re.sub(r"[ \t\n]+", " ", cleaned)  # collapse whitespace
    cleaned = cleaned.rstrip(".,")  # remove trailing punctuation

    # Step 4: try to parse with sympy (with implicit multiplication)
    try:
        expr = parse_expr(
            cleaned,
            transformations=(
                standard_transformations + (implicit_multiplication_application,)
            ),
            evaluate=True,
        )

        # Try simplifying the expression
        expr = simplify(expr)

        # If numeric, return rational string (e.g., "1/2" instead of "0.5")
        if expr.is_number:
            rat = Rational(expr)
            return str(rat) if rat.q != 1 else str(rat.p)

        # Else symbolic
        return str(expr)

    except:
        # Fallback to normalized string
        return cleaned.lower()

def majority_vote(answers):
    """
    Perform majority voting over a list of answers and return the most frequent one.
    """
    # Filter out None and empty strings
    filtered = [a for a in answers if a and a.strip()]
    if not filtered:
        return None

    counts = Counter(filtered)
    max_count = max(counts.values())

    # Get all answers with max frequency
    candidates = [ans for ans, c in counts.items() if c == max_count]

    # Break ties deterministically (lexicographic order)
    return min(candidates)

def plot_figures(chunk_stats, chunk_method):
    """
    - a line plot for x_accuracy, y_accuracy
    - a line plot for x_answer_log_probs_with_eot_correct, y_answer_log_probs_with_eot_correct and x_answer_log_probs_with_eot_incorrect, y_answer_log_probs_with_eot_incorrect, with legened as correct / incorrect
    - a line plot for x_answer_log_probs_without_eot_correct, y_answer_log_probs_without_eot_correct and x_answer_log_probs_without_eot_incorrect, y_answer_log_probs_without_eot_incorrect, with legened as correct / incorrect
    - 2 line plots horizontally next to each other: left: x_answer_log_probs_without_eot_pre, y_answer_log_probs_without_eot_pre (titled "before the first correct answer") right: x_answer_log_probs_without_eot_post, y_answer_log_probs_without_eot_post (titled "after the first correct answer")
    - 2 line plots horizontally next to each other: left: x_answer_log_probs_with_eot_pre, y_answer_log_probs_with_eot_pre (titled "before the first correct answer") right: x_answer_log_probs_with_eot_post, y_answer_log_probs_with_eot_post (titled "after the first correct answer")
    - plots the above 5 figures one after the other vertically (last to containing 2 horizontal subplots)
    - Gives a suptitle `Chunking method: chunk_method`
    """    
    # Extract the data to plot
    x_accuracy, y_accuracy = normalize_arrays(
        [chunk['accuracy'] for chunk in chunk_stats]
    )
    
    x_answer_log_probs_with_eot_correct, y_answer_log_probs_with_eot_correct = normalize_arrays(
        [chunk['answer_log_probs_with_eot'] for chunk in chunk_stats if chunk['accuracy'][-1]]
    )
    x_answer_log_probs_with_eot_incorrect, y_answer_log_probs_with_eot_incorrect = normalize_arrays(
        [chunk['answer_log_probs_with_eot'] for chunk in chunk_stats if not chunk['accuracy'][-1]]
    )
    x_answer_log_probs_with_eot_post, y_answer_log_probs_with_eot_post = normalize_arrays(
        [chunk['answer_log_probs_with_eot'][chunk['accuracy'].index(True):] for chunk in chunk_stats if chunk['accuracy'][-1]]
    )
    x_answer_log_probs_with_eot_pre, y_answer_log_probs_with_eot_pre = normalize_arrays(
        [
            chunk['answer_log_probs_with_eot'][:chunk['accuracy'].index(True)]
            for chunk in chunk_stats
            if chunk['accuracy'][-1] and chunk['answer_log_probs_with_eot'][:chunk['accuracy'].index(True)]
        ]
    )
    
    x_answer_log_probs_without_eot_correct, y_answer_log_probs_without_eot_correct = normalize_arrays(
        [chunk['answer_log_probs_without_eot'] for chunk in chunk_stats if chunk['accuracy'][-1]]
    )
    x_answer_log_probs_without_eot_incorrect, y_answer_log_probs_without_eot_incorrect = normalize_arrays(
        [chunk['answer_log_probs_without_eot'] for chunk in chunk_stats if not chunk['accuracy'][-1]]
    )
    x_answer_log_probs_without_eot_post, y_answer_log_probs_without_eot_post = normalize_arrays(
        [chunk['answer_log_probs_without_eot'][chunk['accuracy'].index(True):] for chunk in chunk_stats if chunk['accuracy'][-1]]
    )
    x_answer_log_probs_without_eot_pre, y_answer_log_probs_without_eot_pre = normalize_arrays(
        [
            chunk['answer_log_probs_without_eot'][:chunk['accuracy'].index(True)]
            for chunk in chunk_stats
            if chunk['accuracy'][-1] and chunk['answer_log_probs_without_eot'][:chunk['accuracy'].index(True)]
        ]
    )

    accuracy_start = [chunk['accuracy'][0] for chunk in chunk_stats]
    accuracy_end = [chunk['accuracy'][-1] for chunk in chunk_stats]
    cot_start = [chunk['chunk_idx'][0] for chunk in chunk_stats]
    cot_end = [chunk['chunk_idx'][-1] for chunk in chunk_stats]

    means_acc = [np.mean(accuracy_start), np.mean(accuracy_end)]
    means_cot = [np.mean(cot_start), np.mean(cot_end)]
    stds_cot = [np.std(cot_start), np.std(cot_end)]
    
    fig = plt.figure(figsize=(12, 16))
    fig.suptitle(f"Chunking method: {chunk_method}", fontsize=16, y=0.95)
    
    gs = gridspec.GridSpec(4, 2, figure=fig)
    
    # 0) Bar plots (row 0, side by side)
    ax0_left = fig.add_subplot(gs[0, 0])
    ax0_left.bar(["First Episode", "Last Episode"], means_acc, color=["steelblue", "seagreen"])
    ax0_left.set_title("Accuracy")
    ax0_left.set_ylabel("Mean")
    
    ax0_right = fig.add_subplot(gs[0, 1])
    ax0_right.bar(["First Episode", "Last Episode"], means_cot, yerr=stds_cot, capsize=5, color=["steelblue", "seagreen"])
    ax0_right.set_title("CoT Length")
    ax0_right.set_ylabel("Mean ± Std")
    
    # 1) Accuracy line plot (row 1, full width)
    ax1 = fig.add_subplot(gs[1, :])
    ax1.plot(x_accuracy, y_accuracy, label="Accuracy")
    ax1.set_title("Accuracy over chunks")
    ax1.set_xlabel("Normalized chunks")
    ax1.set_ylabel("Accuracy")
    ax1.axis(xmin=-0.05, xmax=1.05)
    ax1.legend()
    ax1.grid(True)
    
    # 2) Without EOT correct vs incorrect (row 2, full width)
    ax2 = fig.add_subplot(gs[2, :])
    ax2.plot(x_answer_log_probs_without_eot_correct, y_answer_log_probs_without_eot_correct, label="Correct")
    ax2.plot(x_answer_log_probs_without_eot_incorrect, y_answer_log_probs_without_eot_incorrect, label="Incorrect")
    ax2.set_title("Answer log probs (without EOT)")
    ax2.set_xlabel("Normalized chunks")
    ax2.set_ylabel("Log probs of ANSWER")
    ax2.axis(xmin=-0.05, xmax=1.05)
    ax2.legend()
    ax2.grid(True)
    
    # 3) Without EOT pre vs post (row 3, side by side)
    ax3_left = fig.add_subplot(gs[3, 0])
    ax3_left.plot(x_answer_log_probs_without_eot_pre, y_answer_log_probs_without_eot_pre, label="Pre")
    ax3_left.set_title("Without EOT (before the first correct answer)")
    ax3_left.set_xlabel("Normalized chunks")
    ax3_left.set_ylabel("Log probs of ANSWER")
    ax3_left.axis(xmin=-0.05, xmax=1.05)
    ax3_left.grid(True)
    
    ax3_right = fig.add_subplot(gs[3, 1])
    ax3_right.plot(x_answer_log_probs_without_eot_post, y_answer_log_probs_without_eot_post, label="Post")
    ax3_right.set_title("Without EOT (after the first correct answer)")
    ax3_right.set_xlabel("Normalized chunks")
    ax3_right.set_ylabel("Log probs of ANSWER")
    ax3_right.axis(xmin=-0.05, xmax=1.05)
    ax3_right.grid(True)
    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
    
    # # 5) with_eot pre vs post (side by side)
    # ax5_left = fig.add_subplot(5, 2, 9)
    # ax5_left.plot(x_answer_log_probs_with_eot_pre, y_answer_log_probs_with_eot_pre, label="Pre")
    # ax5_left.set_title("With EOT (before the first correct answer)")
    # ax5_left.set_xlabel("Normalized chunks")
    # ax5_left.set_ylabel("Log probs of </think>ANSWER")
    # ax5_left.grid(True)
    
    # ax5_right = fig.add_subplot(5, 2, 10)
    # ax5_right.plot(x_answer_log_probs_with_eot_post, y_answer_log_probs_with_eot_post, label="Post")
    # ax5_right.set_title("With EOT (after the first correct answer)")
    # ax5_right.set_xlabel("Normalized chunks")
    # ax5_right.set_ylabel("Log probs of </think>ANSWER")
    # ax5_right.grid(True)
    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
```

```python
for chunk_method in ['keywords', 'eot_threshold', 'uniform', 'newlines']:
    chunk_stats = []
    
    for item in chunk_data:
        gt_answer = clean_math_answer(item['answer'])
        
        chunk_idx = []
        accuracy = []
        answer_log_probs_with_eot = []
        answer_log_probs_without_eot = []
        
        for chunk in item['chunks'][chunk_method]:
            predicted_answers = [clean_math_answer(answer) for answer in chunk['decoded_answers']]
            predicted_answer = majority_vote(predicted_answers)
        
            chunk_idx.append(chunk['chunk_end_id'])
            accuracy.append(gt_answer == predicted_answer)
            answer_log_probs_with_eot.append(chunk['answer_log_probs']['with_eot'])
            answer_log_probs_without_eot.append(chunk['answer_log_probs']['without_eot'])
        
        chunk_stats.append(
            {
                'chunk_idx': chunk_idx,
                'accuracy': accuracy,
                'answer_log_probs_with_eot': answer_log_probs_with_eot,
                'answer_log_probs_without_eot': answer_log_probs_without_eot,
            }
        )
    
    plot_figures(chunk_stats=chunk_stats, chunk_method=chunk_method)
```
